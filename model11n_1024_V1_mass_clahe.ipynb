```python
#codi similar al altre notebook per veure les imatges

import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

base = "/content/yolo_cbis/yolo_cbis/"

# 1) Indexar TODOS los .txt de labels en un diccionario {nombre_sin_ext: ruta_completa}
label_root = os.path.join(base, "labels")
label_index = {}

for root, dirs, files in os.walk(label_root):
    for fname in files:
        if fname.lower().endswith(".txt"):
            stem = Path(fname).stem  # nombre sin extensión
            full_path = os.path.join(root, fname)
            label_index[stem] = full_path

print(f"Se han indexat {len(label_index)} fitxers (.txt) en {label_root}")

def load_yolo_labels(label_path):
    boxes = []
    if not os.path.exists(label_path):
        return boxes

    with open(label_path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 5:
                continue
            cls = int(float(parts[0]))
            xc, yc, w, h = map(float, parts[1:5])
            boxes.append((cls, xc, yc, w, h))
    return boxes

def plot_image_with_boxes(img_path, boxes, title=None):
    img = cv2.imread(img_path)
    if img is None:
        print("no es pot llegir la imatge: {img_path}")
        return

    # BGR -> RGB matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w = img.shape[:2]

    for cls, xc, yc, bw, bh in boxes:
        x1 = int((xc - bw / 2) * w)
        y1 = int((yc - bh / 2) * h)
        x2 = int((xc + bw / 2) * w)
        y2 = int((yc + bh / 2) * h)


        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

    plt.figure(figsize=(5, 5))
    plt.imshow(img)
    if title:
        plt.title(title)
    plt.axis("off")
    plt.show()

splits = ["train", "val", "test"]

for split in splits:
    txt_file = os.path.join(base, f"{split}.txt")
    print(f"\n=== Split: {split} ===")
    print(f"Usando lista: {txt_file}")

    if not os.path.exists(txt_file):
        print(f" {txt_file} no existeix")
        continue

    # llegir fitxer
    with open(txt_file, "r") as f:
        lines = [l.strip() for l in f if l.strip()]

    # cogemos 2 imágenes (o menos si no hay tantas)
    sample_imgs = lines[:2]
    print(f"imatges {len(sample_imgs)} ")

    for img_line in sample_imgs:
        # la línea puede ser ruta absoluta o relativa
        if os.path.isabs(img_line):
            img_path = img_line
        else:
            img_path = os.path.join(base, img_line)

        p = Path(img_path)
        stem = p.stem

        # buscar el label correspondiente en el índice
        label_path = label_index.get(stem, None)

        if label_path is None:
            print(f"\nImagen: {img_path}")
            print("nosha trobat el fitxer")
            boxes = []
        else:
            boxes = load_yolo_labels(label_path)
            print(f"\nImagen: {img_path}")
            print(f"Labels: {label_path} -> {len(boxes)} cajas")

        plot_image_with_boxes(
            img_path,
            boxes,
            title=f"{split} - {p.name} ({len(boxes)} cajas)"
        )
```

```python
from ultralytics import YOLO
import ultralytics
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt


data_yaml = "/content/yolo_cbis/yolo_cbis/yolo_cbis.yaml"
MODEL_PATH = "/content/runs/detect/train/weights/best.pt"


experiments = [
    {"name": "exp1", "conf": 0.10, "iou": 0.50},
    {"name": "exp2", "conf": 0.20, "iou": 0.50},
    {"name": "exp3", "conf": 0.30, "iou": 0.50},
    {"name": "exp4", "conf": 0.40, "iou": 0.50},
    {"name": "exp5", "conf": 0.50, "iou": 0.50},
    {"name": "exp6", "conf": 0.60, "iou": 0.50},
    {"name": "exp7", "conf": 0.20, "iou": 0.30},
    {"name": "exp8", "conf": 0.20, "iou": 0.40},
    {"name": "exp9", "conf": 0.20, "iou": 0.45},
    {"name": "exp10", "conf": 0.2, "iou": 0.60},
  ]

# Carregar el model
model = YOLO(MODEL_PATH)

results_paths = []

# Lançar els experiments
for exp in experiments:
    name = exp["name"]
    conf = exp["conf"]
    iou = exp["iou"]

    print(f"\n model {name}  conf={conf}  iou={iou}\n")

    # Avaluació amb val/test split
    metrics = model.val(
        data=data_yaml,
        split="test",
        conf=conf,
        iou=iou,
        save_json=False,
        save_hybrid=False,
        save_txt=False
    )

    save_dir = metrics.save_dir
    results_paths.append((name, save_dir))
```

```python
import shutil

output_folder = "/content/IMAGES_FINALS/"
os.makedirs(output_folder, exist_ok=True)

for name, path in results_paths:
    src = os.path.join(path, "confusion_matrix.png")
    dst = os.path.join(output_folder, f"{name}_confusion.png")
    shutil.copy(src, dst)

print("Guardado en:", output_folder)


```

```python
# codi similar al principal pero ara es poden veure l'imatge real i la pred
import random
base = "/content/yolo_cbis/yolo_cbis/"


N_SAMPLES = 10

# INDEXAR labeled
label_root = os.path.join(base, "labels")
label_index = {}

for root, dirs, files in os.walk(label_root):
    for fname in files:
        if fname.lower().endswith(".txt"):
            stem = Path(fname).stem
            label_index[stem] = os.path.join(root, fname)

print(f"Labels indexats: {len(label_index)} fitxers {label_root}")


def load_labels_by_stem(stem):
    path = label_index.get(stem)
    boxes = []
    if path is None:
        return boxes
    with open(path) as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 5:
                continue
            cls = int(float(parts[0]))
            xc, yc, w, h = map(float, parts[1:5])
            boxes.append((cls, xc, yc, w, h))
    return boxes


def draw_boxes_yolo_norm(img, boxes, color):
    h, w = img.shape[:2]
    img2 = img.copy()
    for cls, xc, yc, bw, bh in boxes:
        x1 = int((xc - bw/2) * w)
        y1 = int((yc - bh/2) * h)
        x2 = int((xc + bw/2) * w)
        y2 = int((yc + bh/2) * h)

        cv2.rectangle(img2, (x1, y1), (x2, y2), color, 4)
    return img2


def draw_boxes_xyxy(img, boxes_xyxy, color):
    img2 = img.copy()
    for cls, x1, y1, x2, y2 in boxes_xyxy:
        cv2.rectangle(img2, (int(x1), int(y1)), (int(x2), int(y2)), color, 4)
    return img2


# Carregar model YOLO
model = YOLO(MODEL_PATH)

# Llegir split de test
txt_file = os.path.join(base, "test.txt")
with open(txt_file, "r") as f:
    img_list = [line.strip() for line in f if line.strip()]

# seleccionar mostres aleatòries
N_SAMPLES = min(N_SAMPLES, len(img_list))
samples = random.sample(img_list, N_SAMPLES)

print(f"\nMostrant i guardant {N_SAMPLES} imatges del test")

# GENERAR IMATGES
for img_rel in samples:
    img_path = img_rel if img_rel.startswith("/") else os.path.join(base, img_rel)
    img_path = os.path.normpath(img_path)
    stem = Path(img_path).stem

    img_bgr = cv2.imread(img_path)
    if img_bgr is None:
        print(f"\nNo s'ha pogut llegir la imatge: {img_path}")
        continue

    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # carregar GT
    gt_boxes = load_labels_by_stem(stem)

    # prediccions YOLO
    pred = model.predict(
        source=img_path,
        imgsz=1024,
        conf=0.1,
        iou=0.4,
        verbose=False
    )[0]

    pred_boxes_xyxy = []
    if pred.boxes is not None and len(pred.boxes) > 0:
        for b in pred.boxes:
            cls = int(b.cls[0])
            x1, y1, x2, y2 = b.xyxy[0].tolist()
            pred_boxes_xyxy.append((cls, x1, y1, x2, y2))

    # dibuixar
    pred_img = draw_boxes_xyxy(img, pred_boxes_xyxy, color=(255, 0, 0))  # vermell
    gt_img   = draw_boxes_yolo_norm(img, gt_boxes,   color=(0, 255, 0))  # verd

    # CREAR FIGURA
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    fig.suptitle(f"{Path(img_path).name}", fontsize=14)

      # subplot 1: Prediccions
    axs[0].set_title("Predicció (vermell)")
    axs[0].imshow(pred_img)
    axs[0].axis("off")

      # subplot 2: Ground Truth
    axs[1].set_title("Reals (verd)")
    axs[1].imshow(gt_img)
    axs[1].axis("off")

      # GUARDAR RESULTAT A DISC
    save_path = os.path.join(output_folder, f"{stem}_comparativa.png")
    fig.savefig(save_path, dpi=300, bbox_inches="tight")

      # (opcional) Mostrar també a la pantalla
    plt.show()

    plt.close(fig)
```
